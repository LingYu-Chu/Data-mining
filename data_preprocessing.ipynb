```python
import numpy as np 
import pandas as pd 

#import data
data = pd.read_csv("diamonds.csv")
```

```python
#information of train data
data.info()
```

```python
#information of each feature
data.describe()
```

# preprocessing

```python
clar_order = ["I1", "SI2", "SI1", "VS2", "VS1", "VVS2", "VVS1", "IF"]
cut_order = ['Fair', 'Good', 'Very Good', 'Premium', 'Ideal']
color_order = sorted(data['color'].unique().tolist(), reverse = True)

#One hot encoding implementation
one_hot_encoders_data =  pd.get_dummies(data)
cols = one_hot_encoders_data.columns
clean_data = pd.DataFrame(one_hot_encoders_data,columns= cols)
clean_data.head()
```

```python
clean_data['cut']=data['cut']
clean_data['color']=data['color']
clean_data['clarity']=data['clarity']
```

```python
#add point and priceperpoint
for dataset in clean_data:
    clean_data['point']=clean_data['carat']*100
    clean_data['priceperpoint']=clean_data['price']/clean_data['point']
    
clean_data.head(3)
```

```python
#cut
Cut = {"Fair": 5, "Good": 4, "Very Good": 3, "Premium": 2, "Ideal": 1}

clean_data=clean_data.replace({"cut": Cut})
clean_data.head(3)
```

```python
#color
Color = {"D": 1,"E": 2, "F": 3, "G": 4, "H": 5, "I": 6, "J": 7}
clean_data=clean_data.replace({"color": Color})
clean_data.head(3)
```

```python
#clarity
Clarity = {"IF" : 1,"VVS1": 2,"VVS2": 3,"VS1": 4, "VS2": 5, "SI1": 6, "SI2": 7, "I1": 8}
clean_data=clean_data.replace({"clarity": Clarity})
clean_data.head(3)
```

# Discretization

```python
#priceperpoint
clean_data.loc[clean_data['priceperpoint'] < 21, 'priceperpoint'] =1
clean_data.loc[(clean_data['priceperpoint'] >= 21) & (clean_data['priceperpoint']<42), 'priceperpoint'] =2
clean_data.loc[(clean_data['priceperpoint'] >= 42) & (clean_data['priceperpoint']<63), 'priceperpoint'] =3
clean_data.loc[clean_data['priceperpoint'] >= 63, 'priceperpoint'] =4
clean_data['priceperpoint']=clean_data['priceperpoint'].astype(int)
clean_data['priceperpoint'].value_counts()
```

```python
#carat
clean_data.loc[clean_data['carat'] < 0.5, 'carat_discrete'] =1
clean_data.loc[(clean_data['carat'] >= 0.5) & (clean_data['carat']<1), 'carat_discrete'] =2
clean_data.loc[clean_data['carat'] >= 1, 'carat_discrete'] =3
clean_data['carat_discrete']=clean_data['carat_discrete'].astype(int)
clean_data['carat_discrete'].value_counts()
```

```python
clean_data.head(3)
```

```python
#drop unused feature: ID, price, table, depth
final_data=clean_data.drop(clean_data.columns[0],axis=1)
final_data=final_data.drop(['x','y','z','price','table','depth','point'],axis=1)
final_data.info()
```
